My dataset: https://www.kaggle.com/datasets/tunguz/inpatient-prospective-payment-system-ipps

Import the necessary libraries and create your dataframe(s).

```python
#import needed libraires with aliases
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Read csv file.
df = pd.read_csv("ipps_cleaned.csv")

# Want to see if the dataframe was properly imported and see what kind of data is stored in each column.
df.info()
df.head()
```

```python
# extracting new column from drg definition column.
df['DRG_definition_without_code']=(df['DRG_Definition'].str.slice(5))
df.info()
```

```python
# exctracting DRG  column from 'DRG_definition_without_code' column
df['DRG']=(df['DRG_definition_without_code'].str.split('W').str.get(0))
df.info()
```

```python
print(df.shape)
df
```

```python
# reading cleaned csv file with no index reference into new dataframe.
df.to_csv("ipps_cleaned.csv",index=False)
df1=pd.read_csv("ipps_cleaned.csv")
df1.info()
```

#  creating  new csv's
1.average covered payments by its discharges in every state of united states.


```python
# create a new dataframe with average covered and medicare  payments .
df_payments= df1[["Provider_State","_Average_Covered_Charges_"]]
```

```python
# create a new data frame with state and the average payments
df_2 = df_payments.groupby("Provider_State")['_Average_Covered_Charges_'].mean() 
```

```python
#save dataset to a new csv
df_2.to_csv("state_covered_payments.csv")
```

```python
# Check to see that it worked
df_st_cov_payments = pd.read_csv("state_covered_payments.csv")

# Want to see if the dataframe was properly imported and see what kind of data is stored in each column.
df_st_cov_payments.info()
df_st_cov_payments
```

we have noticed  average covered charges rates are highest in the  state of California.

## Total discharges by   indivivdual DRG through country.

```python
# creting dataframe applying group by function and calculating total discharges by each DRG group.
df_discharges=df[["DRG","_Total_Discharges_" ]]
df3=df_discharges.groupby("DRG")["_Total_Discharges_"].sum()
df3.to_csv("Discharges_DRG.csv")
df_dis_D=pd.read_csv("Discharges_DRG.csv")
```

```python
# printing new dataframe which is  group by drg definition .
df_dis_D.info()
df_dis_D.head(57)
```

## percentage of medicare coverage 

```python
# finding medicare covered percentage.
med_cov_perc= ((df["Average_Medicare_Payments"]*100)/df["_Average_Total_Payments_"])
print("medicare covered percentage:",med_cov_perc)
```

## revenue from medicare pay.


```python
# revenue from medicare payments.
revenue_med_pay=round(df["_Total_Discharges_"]*df["Average_Medicare_Payments"])
print("revenue of medicare: ",revenue_med_pay)
revenue_med_pay.describe()
```

Here with manuplation we have obsereved that california has highest coverage of medicare payments..further with visualization we foucse on states with highest coverage and drg has most better  medicare coverage across the usa.
